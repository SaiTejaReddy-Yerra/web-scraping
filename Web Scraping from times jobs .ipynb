```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://m.timesjobs.com/jobfunction/it-software-jobs"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36"}

page = requests.get(url, headers = headers)
soup = BeautifulSoup(page.text, 'html.parser')
jobs = soup.find('ul', class_ ="ui-content search-result")
job_link_tag = job.find('a')
job_url = job_link_tag["href"]


job_response = requests.get(job_url, headers = headers)
job_soup = BeautifulSoup(job_response.text, 'html.parser')

job_title = job_soup.find('h1').text.strip()
job_location = job_soup.find('div',class_ = 'srp-loc jd-loc').text.strip()
salary_range = job_soup.find('div',class_ = 'srp-sal').text.strip()
post_date = job_soup.find('span',class_ = 'posting-time').text.strip()

print(post_date)
```

```python
url = "https://m.timesjobs.com/jobfunction/it-software-jobs"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36"}

page = requests.get(url, headers = headers)
soup = BeautifulSoup(page.text, 'html.parser')

jobs = soup.find_all('li')

try:
    for job in jobs:
        job_link_tag = job.find('a')
        job_url = job_link_tag["href"]

        job_response = requests.get(job_url, headers = headers)
        job_soup = BeautifulSoup(job_response.text, 'html.parser')

        job_title = job_soup.find('h1').text.strip()
        job_locations = job_soup.find('div',class_ = 'srp-loc jd-loc').text.strip().split(':')[1]
        salary_range = job_soup.find('div',class_ = 'srp-sal').text.strip()
        post_date = job_soup.find('span',class_ = 'posting-time').text.strip()

        print(f'''
        Job Title : {job_title}
        Job Location :{job_location}
        Salary Range :{salary_range}
        ''')
        
        print()
        
except TypeError:
    print("error")
```
